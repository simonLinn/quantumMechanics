# Numerically Solving the Infinite Quantum Well
<br>
<ul>
<h1>Project#1</h1>
<h3><li>Consider an $1D$ Infinite Quantum Well : </li>

$$
V(x)=
\begin{cases} 
0,& 0 \leq x \leq L \\
\infty,& x<0 , x>L
\end{cases}
 $$



 $$L = 5 nm = 50 \mathring{A}  ,  \Delta x = 2.5 \mathring{A} $$</ul></h3>
<ul>
<h3><li>time independent Schrödinger Equation(TISE)： </li></h3>
<h3>$$ -\frac{\hbar^2}{2m}\frac{d^2}{d x^2} \psi(x) + V(x)\psi(x) = E\psi(x)$$</h3>
</ul>    

<ul>
<h3><li>Discretization of TISE:</li></h3>
<h3> Turn continuous space $x$ into a discrete number of steps , $N$ .</h3>
<h3>Than get $N$ values $x_i$ , $h = \Delta x = x_{i+1} - x_i$ .</h3>
<h3>Assume $\Delta x$ is a constant and small enough.</h3>
<h3>$$
\frac{d \psi(x)}{dx} = \lim_{\Delta x \rightarrow 0} \frac{\psi(x+\Delta x) - \psi(x)}{\Delta x} \approx \frac{\psi(x+h) - \psi(x)}{x_{i+1} - x_i} 
$$</h3>
    
<h3>
  $$
\frac{d^2 \psi(x)}{dx^2} = \frac{d}{dx} \frac{d \psi(x)}{dx} = \frac {\frac{\psi(x_{i+1})-\psi(x_i)}{\Delta x}-\frac{\psi(x_i)-\psi(x_{i-1})}{\Delta x} }{\Delta x}
$$
<br>
$$=\frac{\psi(x_{i+1})-2\psi(x_i)+\psi(x_{i-1})}{\Delta x^2}$$
</h3>

<h3>
So the TISE can be discretized as ：</h3><br>     
<h3>
$$
-\frac{\hbar^2}{2m} \frac{\psi(x_{i+1})-2\psi(x_i)+\psi(x_{i-1})}{\Delta x^2} + 0 * \psi(x)= E\psi(x)$$
</h3>
<h3>
$$\Longrightarrow
\frac{-\hbar^2}{2 m \Delta x^2} \begin{pmatrix} -2 & 1 & 0 & \ddots & \ddots& 0 \\ 1 & -2 & 1 & 0 & \ddots& 0 \\ 
0& 1 & -2 & 1 & \ddots& 0 \\\ddots & \ddots & \ddots & \ddots & 1 & 0\\
0 & \ddots& \ddots& \ddots & 1 & -2 \end{pmatrix} \begin{pmatrix} \psi_1 \\ \psi_2 \\ \ddots \\ \ddots \\ \psi_{N-1}\end{pmatrix} = E \begin{pmatrix} \psi_1 \\ \psi_2 \\ \ddots \\ \ddots \\ \psi_{N-1}\end{pmatrix}
$$ 

</h3>
<h3>
After discretization , TISE becomes an eigenvalues promble , and eigenvalues $E$ the eigenvectors $\vec{\psi}$ can be solved .
    
</h3>
</ul>    

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as scl
HBAR = 6.62607004 * (10**-34)
N = 21
L = 5. * (10**-9)
DeltaX=2.5 * (10**-10)
M=9.10938356*  (10**-31)
squareMatrix=(np.diag(np.ones(N-1),-1) -2* np.diag(np.ones(N),0) + np.diag(np.ones(N-1),1))
x = np.linspace(0.,L,N)

H=np.true_divide(-1*np.power(HBAR,2),(2*M*np.power(DeltaX,2)))*squareMatrix
E,psi = np.linalg.eigh(H)
psi=psi.T
```

```python
plt.figure(figsize=(15,10))
for i in range(5):
   plt.plot(x,psi[i]/np.sqrt(DeltaX),label="$E_{}$={:>0.30f}".format(i,E[i]))
plt.title("Project#1")
plt.legend()
plt.show()
```
